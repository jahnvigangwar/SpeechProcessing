```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
```

```python
audio_file = 'audio_file.wav'
y, sr = librosa.load("C:\\Users\\acer\\Downloads\\sample2.wav", sr=None, mono=True)
```

```python
fft_result = np.fft.fft(y)
frequencies = np.fft.fftfreq(len(y), d=1/sr)

# Plot the magnitude of the Fourier Transform
plt.figure(figsize=(10, 6))
plt.plot(frequencies, np.abs(fft_result))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Magnitude Spectrum')
plt.grid()
plt.show()
```

```python
spectral_centroid = librosa.feature.spectral_centroid(y=y, sr=sr)[0]

# Calculate the Spectral Bandwidth
spectral_bandwidth = librosa.feature.spectral_bandwidth(y=y, sr=sr)[0]

# Calculate the MFCCs
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

# Print the features
print(f"Spectral Centroid: {spectral_centroid}")
print(f"Spectral Bandwidth: {spectral_bandwidth}")
print(f"MFCCs:\n{mfccs}")
```
